
## üìä Step 5: Grouping and Aggregating

Functions to explore:

```python

df.groupby("Sex")["Survived"].mean()
# El metodo groupBy agrupa por columnas, agrupa las filas segun el valor del argumento
# y la segunda, es una lista con los nombres de las columnas, y las selecciona el con la funcion que venga
# por ultimo se hace una media de esa columna dentro del groupBy
```

```python
df["Fare"].mean()
#Devuelve la media de la columna Fare asi como su tipo
```

```python
df["Age"].median()
#Devuelve la mediana de la columna "Age" en este caso no devuleve el tipo
```


## ‚öôÔ∏è Step 7: Creating or Modifying Columns

Functions to explore:
- df.assign()
- df.apply()
- df["new_col"] = ...
- pd.concat()
- pd.merge()


```python
df["Fare_per_Age"] = df["Fare"] / df["Age"]

#En este caso se crea una serie nueva con el nombre entre [ ] en este caso utiliznado la columna "Fare" 
# y "Age" del dataframe
```

```python
df[["Age", "Fare", "Fare_per_Age"]].head()
#En este caso devuelve un data frame con las columas escogidas, pasadas un diccionario para especificar las columnas a ver  las primeras 5
```

```python
df.assign()
#Retorna un nuevo objeto con todas las columnas nuevas que se tengan. 
#Las columnas existentes serian sobreescritos
#Los argumentos que se les pasa son la columna y el valor nuevo que se le va a agregar al dataframe

#Si no se le pasa el argumneto te devulve la misma tabla
```

```python
data1= df.apply('Survived')
data1
#Aplica a cada columna o fila con el axis, dentro de los argumentos, se puede aplicar un funcion para filtrar como un minimo, maximo, media....
# donde f seria la funcion que se pasa
```

```python
df_final = pd.concat([df, data1], axis=1)
df_final
#Devuelve el dataframe que resulta de apilar o unir objetos en panda(Series o Dataframe), utilizando un axis, por default filas y con 1 columnas
```

```python
df_merged = pd.merge(df, data1, on='Survived', how='left')
df_merged
#Deveulve el dataframe que resulta de mezclar 2, un df1 y otro df2(pasados por argumentos )
#Usando como claves las columnas de la lista de clave siguiendo el metodo de tipo(argumento how = tipo) inner, left, right, outer 
```
